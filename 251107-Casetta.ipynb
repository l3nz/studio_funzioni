```python
def to_latex(a):
    latex = "\\begin{align}\n"
    for (n, d) in enumerate(a): 
        if isinstance(d, dict):
            for i, (key, eq) in enumerate(d.items(), 1):
                latex += f"{sp.latex(eq)} \\quad &\\text{{({n, i})}} \\\\\n"
        else:
            latex += f"{sp.latex(d)} \\quad &\\text{{({n})}} \\\\\n"
    latex = latex.rstrip(" \\\\\n") + "\n\\end{align}"
    return Math(latex)
```

## Esercizio 4: La casetta

     |<    r      >|<  x  >|
     
     +---------------------+
     |            /\       |
     |    C     /    \  A  |
     |        /        \   |
     |      /            \ |
     |    /               \|
     |  /                 /|
     |/        B        /  |
     |\               /    |
     |  \           /      |
     |    \       /        |
     |  D   \   /      E   |
     |       \ /           |
     +---------------------+

La casetta è formata dall'area A+B, inscritta in un quadrato di lato 7 cm.
Si cerca x data l'area totale di 16 cm2.

Nota che l'area di A e D è la medesima, così come C ed E.
     

```python
x, r, aa, ac = sp.symbols("x r aa ac")

# r è il lato del triangolo C
r = 7 - x

# area di A ed area di C
aa = x**2 / 2
ac = r**2 / 2

```

La casetta è quanto rimane dopo aver tolto all'area totale (7*7) D, C ed E, dove C=E e D=A

```python
fnab = 49 - (ac*2 + aa)
```

```python
to_latex([fnab])
```

```python
sp.expand(fnab-16)
```

Risolvo per fnab=16

```python
to_latex(sp.solve(fnab-16))
```

Qui vedo che la soluzione (1) è fuori del dominio ammesso per x, quindi è valida la (0).

Volendo potrei aggiungerlo come vincolo al sistema:

```python
sys = [
    fnab - 16,
    x <= 7
]
to_latex(sys)
```

```python
sp.solve(sys)
```

## Modo diretto: calcolo della casetta

```python
ia, ic = sp.symbols("ia ic")

ia = sp.sqrt(x*x + x*x)
ic = sp.sqrt(r**2 + r**2)
```

```python
to_latex([ic])
```

```python
fnac = aa + (ia*ic) - 16
```

```python
to_latex(sp.solve(fnac))
```

Dove la soluzione (1) non è accettabile in quanto > 7
